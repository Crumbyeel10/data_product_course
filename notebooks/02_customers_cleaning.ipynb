```python
import pandas as pd

customers = pd.read_csv('../data_product_course/data/raw/customers.csv')
customers.head()
```

```python
import pandas as pd

df = customers.copy()

df = customers.copy()

# 1) Trim de texto
df["name"] = df["name"].astype(str).str.strip()

# 2) Email: minúsculas + trim
df["email"] = df["email"].astype(str).str.strip().str.lower()

# 3) Fecha: convertir a datetime
df["created_at"] = pd.to_datetime(df["created_at"], errors="coerce")

df.head()
```

```python
# customer_id no nulo
null_customer_id = df["customer_id"].isna().sum()

# customer_id único
dup_customer_id = df.duplicated(subset=["customer_id"]).sum()

# created_at parseado (no NaT)
null_created_at = df["created_at"].isna().sum()

print("Null customer_id:", null_customer_id)
print("Duplicate customer_id:", dup_customer_id)
print("Null created_at (parse errors):", null_created_at)

assert null_customer_id == 0, "Hay customer_id nulos"
assert dup_customer_id == 0, "Hay customer_id duplicados"
assert null_created_at == 0, "Hay created_at inválidos (NaT)"
print("✅ Quality checks OK")
```

```python
output_path = "../data_product_course/data/processed/silver_customers.csv"
df.to_csv(output_path, index=False)
print("Guardado:", output_path)
```
